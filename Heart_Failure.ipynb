```python
from scipy import stats
survived=data[data["DEATH_EVENT"]==0]["ejection_fraction"]
not_survived=data[data["DEATH_EVENT"]==1]["ejection_fraction"]
t_statistic,p_value=stats.ttest_ind(survived,not_survived)
print("T-statistic:",t_statistic)
print("P-value:",p_value)
if p_value < 0.05:
    print("There is a statistically significant difference between the mean ejection fractions of the two groups.")
else:
    print("There is no statistically significant difference between the mean ejection fractions of the two groups.")
```

```python
contingency_table_smoking = pd.crosstab(data['smoking'], data['DEATH_EVENT'])
chi2, p, dof, expected = stats.chi2_contingency(contingency_table_smoking)
print("Chi-squared statistic:", chi2)
print("P-value:", p)
print("Degrees of freedom:", dof)
print("Expected frequencies:\n", expected)
if p < 0.05:
    print("There is a statistically significant association between smoking and death event.")
else:
    print("There is no statistically significant association between smoking and death event.")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
features = data.drop(columns=['DEATH_EVENT'])
target = data['DEATH_EVENT']
scaled_features = StandardScaler().fit_transform(features)
pca=PCA(n_components=7)

pca_result=pca.fit_transform(scaled_features)

pca_df=pd.DataFrame(data=pca_result,columns=["PC1","PC2","PC3","PC4","PC5","PC6","PC7"])
pca_df['DEATH_EVENT'] = target
pca_df
```

```python
explained_variance=pca.explained_variance_ratio_
explained_variance
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o', linestyle='--')
plt.title('Explained Variance by Principal Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.show()
```

```python
cumulative_variance = explained_variance.cumsum()
cumulative_variance

```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='--', color='red')
plt.title('Cumulative Explained Variance by Principal Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="PC1", y='PC2',data=pca_df, hue=target, palette='viridis')
plt.title('PCA - First Two Principal Components')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()
```
